```python
import pandas as pd
```

```python
df = pd.read_csv('pokemon_data.csv')
```

```python
#Wie viele Zeilen in df?
len(df)
```

```python
#print(df)

#Gibt alle Spalten des Dataframes aus
print(df.columns)
```

```python
#Liefert eine gute Übersicht über das DF
#Hier z.B. zu erkennen: 
#DF hat 800 Zeilen, 
#Ein Pokemon hat durchschnittlich 69 hp,
#der stärkste Angriff ist 90
df.describe()
```

```python
#Zeigt die ersten Zeilen an, 
#Anzahl der Zeilen kann als Parameter übergeben werden
#Standardmäßig die ersten 5
print(df.head(3))
```

```python
#Es können auch mehrere Columns ausgewählt werden, 
#indem eine Liste der Columns erstellt 
#Wichtig: Doppelte eckige Klammern nicht vergessen
print(df[['Name','HP']])
```

```python
#iloc = Integer Location, gewünschte Zeilen und
#gewünschte Spalten können mit int angegeben
#print(df.iloc[0:4])
#5 Zeilen, Spalten von 2-3
#df.iloc[:5,1:3]
#Gewünschte Spalten oder Zeilen können auch als Liste angegeben werden
df.iloc[:,[1,4,5,6]]
```

```python
df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed']
```

```python
df.sort_values('Total')
```

```python
df.drop(columns=['Total'])
```

```python
#andere Möglichkeit eine neue Total Spalte zu erstellen
#axis = 1: Summe horizontal erstellen, axis ist standardmäßig = 0
df['Total'] = df.iloc[:,4:10].sum(axis=1)

#Erstellt eine Liste mit den Spaltennamen
cols = list(df.columns)

#DF Spalten umformen:
#Liste erstellen mit Spalten, 
#cols[-1] muss mit eckigen Klammern umschlossen werden, 
#damit es als Liste erkannt wird, ansonsten String Objekt

df = df[cols[0:4] + [cols[-1]] + cols[4:12]]
```

```python
#Daten in neuer Datei speichern
#Index rauslassen, standardmäßig True
df.to_csv('pokemon_data_modified.csv', index=False)
df.to_excel('pokemon_data_modified.xlsx', index=False)
```

Statistische Funktionen in Pandas

```python
#Mittelwert
df['HP'].mean()
```

```python
#Varianz
#"Im Durchschnitt weicht das Quadrat der HP 652 Punkte vom Mittelwert ab"
df['HP'].var()
```

```python
#Standardabweichung
#"Durchschnittlich liegt ein HP-Wert 25.5 Punkte vom Mittelwert entfernt"
df['HP'].std()
```

```python
#75% der Werte liegen über 50 HP, 25% der Werte liegen unter 50 HP
df['HP'].quantile(0.25)
```

```python
df['HP'].quantile(0.75)
```

```python
#| ist ein Oder, alternativ lässt sich & verwenden, wenn beide Werte gewünscht sind, erzeugt Kopie von df
df[(df['HP'] > 50) | (df['HP'] == 1) ][['Name','HP']]
```

```python
#Spalte löschen, nicht auf Kopie sondern auf df
df.drop(['Legendary'], axis = 1, inplace = True )
```
