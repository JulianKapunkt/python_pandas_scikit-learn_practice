```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
#Zeigt Diagramme in Jupyter an
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
#plt.show() in Pycharm zum anzeigen
```

```python
#(nrows, nrcols, plt nr.)
plt.subplot(1,2,1)
plt.plot(x,y, 'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```
